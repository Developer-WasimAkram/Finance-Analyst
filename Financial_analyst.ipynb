## 2. Data Cleaning — Fixing Messy Data
Before any analysis, a financial analyst must clean the data: find missing values, remove duplicates, and fix data types.

```python
"""
COUNT MISSING VALUES — .isnull().sum()
- .isnull() returns True/False for each cell (True = missing/NaN).
- .sum() counts the True values per column (since True=1, False=0).
- .sum().sum() gives the total across ALL columns.
- Missing data leads to wrong analysis — always check and handle it.
- In finance, missing data could mean unreported transactions or incomplete records.
"""
print("Missing values per column:\n")
print(data.isnull().sum())
print(f"\nTotal missing values: {data.isnull().sum().sum()}")
```

```python
"""
MISSING VALUE PERCENTAGE — .isnull().mean()
- .isnull().mean() calculates the fraction of missing values per column (0.0 to 1.0).
- Multiply by 100 to get percentage.
- Rule of thumb for handling missing data:
  - <5% missing: safe to fill with mean/median/mode
  - 5-30% missing: fill carefully or use advanced imputation
  - >30% missing: consider dropping the column entirely
"""
print("Missing value percentage per column:\n")
print((data.isnull().mean() * 100).round(2))
```

```python
"""
FIND & REMOVE DUPLICATES — .duplicated(), .drop_duplicates()
- .duplicated() returns True for rows that are exact copies of an earlier row.
- .duplicated().sum() counts total duplicate rows.
- .drop_duplicates() removes duplicate rows, keeping the first occurrence.
- In financial data, duplicates could mean double-counted transactions — always check!
- You can check duplicates on specific columns: data.duplicated(subset=['id'])
"""
print(f"Duplicate rows: {data.duplicated().sum()}")
data_clean = data.drop_duplicates()
print(f"Rows after removing duplicates: {len(data_clean)}")
```

```python
"""
FILL MISSING VALUES — .fillna()
- .fillna(value) replaces all NaN (missing) cells with the specified value.
- Common strategies:
  - Fill with 0: for numeric columns where missing means "none" (e.g., no reviews = 0)
  - Fill with a string: like 'Unknown' or 'No Review' for text columns
  - Fill with mean/median: data['col'].fillna(data['col'].mean()) for numeric columns
- Always think about WHAT makes sense for each column — don't blindly fill everything with 0.
- After filling, verify with .isnull().sum() that missing values are handled.
"""
data['reviews_per_month'] = data['reviews_per_month'].fillna(0)
data['last_review'] = data['last_review'].fillna('No Review')
data['host_name'] = data['host_name'].fillna('Unknown')
print("Missing values after cleaning:\n", data.isnull().sum())
```

```python
"""
CONVERT DATA TYPES & COUNT UNIQUES — .astype(), pd.to_datetime(), .nunique()
- pd.to_datetime() converts text dates like '2024-06-09' into proper date objects.
- errors='coerce' means: if a value can't be converted, make it NaT (Not a Time) instead of crashing.
- .astype() converts a column to a different type: data['col'].astype(int), .astype(str), etc.
- .nunique() counts the number of unique/distinct values in each column.
- Useful for identifying: IDs (high unique count), categories (low unique count), and constants (1 unique).
"""
data['last_review'] = pd.to_datetime(data['last_review'], errors='coerce')
print("Unique values per column:\n")
print(data.nunique())
```

## 3. Statistical Analysis — Numbers That Tell a Story
A financial analyst lives by statistics. These methods summarize thousands of rows into actionable insights.

```python
"""
CENTRAL TENDENCY & SPREAD — .mean(), .median(), .mode(), .std(), .var(), .min(), .max()
- mean: the average (sum of all values / count) — sensitive to outliers
- median: the middle value when sorted — better for skewed data like prices
- mode: the most frequently occurring value
- std (standard deviation): how far values typically deviate from the mean
- var (variance): std squared — used in financial risk calculations
- Range (max - min): the total spread of values
- As a financial analyst, compare mean vs median: if mean >> median, you have outliers pulling the average up.
"""
print("=== Price Analysis ===")
print(f"Mean (average) price:   ${data['price'].mean():.2f}")
print(f"Median (middle) price:  ${data['price'].median():.2f}")
print(f"Mode (most common):     ${data['price'].mode()[0]}")
print(f"Std deviation (spread): ${data['price'].std():.2f}")
print(f"Variance:               ${data['price'].var():.2f}")
print(f"Min price:              ${data['price'].min()}")
print(f"Max price:              ${data['price'].max()}")
print(f"Range:                  ${data['price'].max() - data['price'].min()}")
```

```python
"""
PERCENTILES, SKEWNESS & KURTOSIS — .quantile(), .skew(), .kurtosis()
- .quantile(q) returns the value below which q% of data falls.
  - 25th percentile = 25% of listings cost less than this
  - 90th percentile = only 10% of listings cost MORE than this
- .skew() measures asymmetry of the distribution:
  - 0 = perfectly symmetric (bell curve)
  - >0 = right-skewed (long tail to the right — many cheap, few expensive)
  - <0 = left-skewed (long tail to the left)
- .kurtosis() measures "tailedness" / extreme values:
  - High kurtosis = more extreme outliers
  - Low kurtosis = fewer outliers
- Financial analysts use percentiles for risk thresholds (e.g., Value at Risk).
"""
print("=== Price Percentiles ===")
for q in [0.10, 0.25, 0.50, 0.75, 0.90, 0.95, 0.99]:
    print(f"  {int(q*100)}th percentile: ${data['price'].quantile(q):.0f}")
print(f"\nSkewness: {data['price'].skew():.2f}")
print(f"Kurtosis: {data['price'].kurtosis():.2f}")
```

```python
"""
CORRELATION MATRIX — .corr()
- .corr() calculates the Pearson correlation coefficient between every pair of numeric columns.
- Values range from -1 to +1:
  - +1 = perfect positive correlation (when one goes up, the other goes up)
  - -1 = perfect negative correlation (when one goes up, the other goes down)
  - 0 = no linear relationship
- .select_dtypes(include='number') filters only numeric columns (excludes text/dates).
- In finance, correlation helps identify: diversification opportunities, risk factors, and pricing drivers.
- Example: if price and reviews are positively correlated, popular listings may charge more.
"""
numeric_cols = data.select_dtypes(include='number')
correlation = numeric_cols.corr()
print("=== Correlation Matrix ===\n")
correlation.round(2)
```

```python
"""
FREQUENCY DISTRIBUTION — .value_counts()
- .value_counts() counts how many times each unique value appears in a column.
- Returns results sorted by frequency (most common first).
- .head(10) limits output to top 10.
- Use normalize=True to get percentages: data['col'].value_counts(normalize=True)
- Essential for understanding categorical data: What room types dominate? Which neighborhoods have most listings?
"""
print("=== Room Type Distribution ===")
print(data['room_type'].value_counts())
print("\n=== Top 10 Neighbourhoods ===")
print(data['neighbourhood'].value_counts().head(10))
```

## 4. Grouping & Aggregation — Segment Analysis
Financial analysts always break data into segments. GroupBy is the most powerful pandas tool for this — it splits data into groups, applies a function, and combines results.

```python
"""
GROUP BY WITH SINGLE AGGREGATION — .groupby() + .mean()
- .groupby('column') splits the DataFrame into groups based on unique values in that column.
- Then you apply an aggregation function like .mean(), .sum(), .count(), .median(), etc.
- This is the Split-Apply-Combine pattern:
  1. SPLIT: divide data by room_type (Entire home, Private room, etc.)
  2. APPLY: calculate mean price for each group
  3. COMBINE: merge results into a single table
- This answers: "What is the average price for each room type?"
"""
print("=== Average Price by Room Type ===")
data.groupby('room_type')['price'].mean().round(2)
```

```python
"""
MULTIPLE AGGREGATIONS — .groupby() + .agg()
- .agg() lets you calculate MULTIPLE statistics in one call.
- Syntax: .agg(new_name='function_name') gives custom column names.
- Available functions: 'count', 'mean', 'median', 'min', 'max', 'std', 'sum', 'var', 'first', 'last'
- This is the analyst's Swiss army knife — one line gives you a complete statistical profile per group.
- Think of it as: "For each neighbourhood group, give me the count, average, median, min, max, and std of prices."
"""
print("=== Neighbourhood Group Analysis ===")
data.groupby('neighbourhood_group')['price'].agg(
    count='count',
    mean_price='mean',
    median_price='median',
    min_price='min',
    max_price='max',
    std_dev='std'
).round(2)
```

```python
"""
MULTI-LEVEL GROUPBY — .groupby() with multiple columns + .unstack()
- Pass a list of columns to groupby: groupby(['col1', 'col2'])
- This creates a hierarchical grouping (groups within groups).
- .unstack() pivots the inner group into columns for easier reading.
- Example: average price for each room type WITHIN each neighbourhood group.
- Useful for comparing how a metric varies across two dimensions simultaneously.
"""
print("=== Price by Neighbourhood Group & Room Type ===")
data.groupby(['neighbourhood_group', 'room_type'])['price'].mean().round(2).unstack()
```

```python
"""
PIVOT TABLE — pd.pivot_table()
- pivot_table() works like Excel's Pivot Tables — summarize data by two dimensions.
- Parameters:
  - values: the column to aggregate (what to calculate)
  - index: row labels (groups on the left side)
  - columns: column labels (groups across the top)
  - aggfunc: function(s) to apply — can be a list like ['mean', 'count']
- This creates a summary table showing mean price AND count for each area/room type combination.
- Financial analysts use pivot tables daily for segment reporting.
"""
pivot = pd.pivot_table(
    data,
    values='price',
    index='neighbourhood_group',
    columns='room_type',
    aggfunc=['mean', 'count']
).round(2)
print("=== Pivot Table: Price by Area & Room Type ===")
pivot
```

```python
"""
CROSS-TABULATION — pd.crosstab()
- pd.crosstab() counts the number of occurrences for each combination of two categorical columns.
- margins=True adds row and column totals (like Excel's grand total).
- Similar to pivot_table with aggfunc='count', but simpler syntax for counting.
- Useful for answering: "How many of each room type are in each neighbourhood group?"
- You can add normalize='index' to get row percentages instead of counts.
"""
print("=== Cross Tab: Room Type vs Neighbourhood Group ===")
pd.crosstab(data['neighbourhood_group'], data['room_type'], margins=True)
```

## 5. Filtering, Sorting & Selection — Finding What Matters
Financial analysts need to quickly slice data to answer specific questions. These are your data scalpel tools.

```python
"""
BOOLEAN FILTERING — Selecting rows by condition
- data[condition] returns only rows where the condition is True.
- Use comparison operators: < > <= >= == !=
- Combine conditions with & (AND) and | (OR) — each condition MUST be in parentheses.
- This is how you answer: "Show me all listings under $100" or "Find cheap entire homes with good reviews."
- The result is a new DataFrame containing only matching rows.
"""
budget_listings = data[data['price'] < 100]
print(f"Budget listings (under $100): {len(budget_listings)}")

# Multiple conditions with & (AND) and | (OR)
deals = data[(data['price'] < 100) & (data['room_type'] == 'Entire home/apt') & (data['number_of_reviews'] > 50)]
print(f"Great deals (cheap + entire home + 50+ reviews): {len(deals)}")
deals[['name', 'neighbourhood', 'price', 'number_of_reviews']].head(10)
```

```python
"""
SORTING DATA — .sort_values()
- .sort_values('column') sorts the DataFrame by a column.
- ascending=False sorts from highest to lowest (descending order).
- ascending=True (default) sorts from lowest to highest.
- Chain with .head(n) to get "Top N" results.
- You can sort by multiple columns: .sort_values(['col1', 'col2'], ascending=[False, True])
"""
print("=== Top 10 Most Expensive Listings ===")
data.sort_values('price', ascending=False).head(10)[['name', 'neighbourhood_group', 'room_type', 'price']]
```

```python
"""
TOP/BOTTOM N ROWS — .nlargest(), .nsmallest()
- .nlargest(n, 'column') returns the n rows with the LARGEST values in that column.
- .nsmallest(n, 'column') returns the n rows with the SMALLEST values.
- Faster and cleaner than .sort_values().head() for getting top/bottom records.
- Great for: "Who are the top 5 most reviewed listings?" or "What are the cheapest options?"
"""
print("=== Top 5 Most Reviewed Listings ===")
data.nlargest(5, 'number_of_reviews')[['name', 'neighbourhood', 'price', 'number_of_reviews']]
```

```python
"""
FILTER BY LIST — .isin()
- .isin(list) returns True for rows where the column value is in the given list.
- Much cleaner than writing: (col == 'A') | (col == 'B') | (col == 'C')
- Use ~data['col'].isin(list) to get rows NOT in the list (the ~ means NOT).
- Here we focus on 4 specific neighbourhoods and get their price statistics.
"""
target_areas = ['Capitol Hill', 'Ballard', 'Fremont', 'Wallingford']
area_data = data[data['neighbourhood'].isin(target_areas)]
print(f"Listings in target areas: {len(area_data)}")
area_data.groupby('neighbourhood')['price'].describe().round(2)
```

```python
"""
RANGE FILTER & INDEXING — .between(), .loc[], .iloc[]
- .between(low, high) returns True for values within the range (inclusive on both ends).
- .loc[rows, columns] selects by LABEL (column names and row index labels).
- .iloc[rows, columns] selects by INTEGER POSITION (0-based).
- Use .loc when you know column names; use .iloc when you know positions.
- Example: data.iloc[0:5, 0:3] = first 5 rows, first 3 columns.
"""
mid_range = data[data['price'].between(75, 200)]
print(f"Mid-range listings ($75-$200): {len(mid_range)} ({len(mid_range)/len(data)*100:.1f}% of all listings)")

print("\n=== .loc[] example: First 3 rows, specific columns ===")
data.loc[0:2, ['name', 'price', 'room_type']]
```

```python
"""
SQL-LIKE FILTERING — .query()
- .query('expression') filters rows using a string expression — like writing SQL WHERE clauses.
- More readable than bracket notation for complex conditions.
- Use 'and', 'or', 'not' (or &, |, ~) inside the string.
- Column names can be used directly without data['col'] syntax.
- You can reference external variables with @: data.query('price > @max_price')
"""
expensive_popular = data.query('price > 200 and number_of_reviews > 100')
print(f"Expensive & popular listings: {len(expensive_popular)}")
expensive_popular[['name', 'neighbourhood', 'price', 'number_of_reviews']].head(10)
```

## 6. Financial Analyst Metrics — Calculated Columns & Revenue Estimation
Create new columns with derived financial metrics. This is where raw data turns into business intelligence.

```python
"""
CALCULATED COLUMNS — Creating new financial metrics
- You can create new columns by assigning a formula: data['new_col'] = expression.
- Here we calculate:
  - occupancy_rate: percentage of the year the listing is booked (higher = more popular)
  - estimated_booked_nights: total nights booked per year
  - estimated_annual_revenue: price x booked nights (rough yearly income for the host)
  - estimated_monthly_revenue: annual revenue / 12
- These derived metrics turn raw data into actionable business intelligence.
- Real analysts would also factor in cleaning fees, service fees, and seasonal variation.
"""
data['occupancy_rate'] = ((365 - data['availability_365']) / 365 * 100).round(2)
data['estimated_booked_nights'] = 365 - data['availability_365']
data['estimated_annual_revenue'] = data['price'] * data['estimated_booked_nights']
data['estimated_monthly_revenue'] = (data['estimated_annual_revenue'] / 12).round(2)

print("=== Revenue Estimates (Top 10 Earners) ===")
data.nlargest(10, 'estimated_annual_revenue')[
    ['name', 'neighbourhood', 'price', 'occupancy_rate', 'estimated_annual_revenue']
]
```

```python
"""
REVENUE ANALYSIS BY SEGMENT — .groupby() + .agg() with multiple columns
- Combine groupby with agg to create a comprehensive financial summary per segment.
- Uses named aggregation: new_column_name=('source_column', 'function').
- This creates a mini financial report for each neighbourhood group showing:
  - Total listings (market size)
  - Average price (pricing benchmark)
  - Average occupancy (demand indicator)
  - Total revenue (market value)
  - Average revenue per listing (profitability benchmark)
"""
print("=== Revenue Summary by Neighbourhood Group ===")
revenue_by_area = data.groupby('neighbourhood_group').agg(
    total_listings=('id', 'count'),
    avg_price=('price', 'mean'),
    avg_occupancy=('occupancy_rate', 'mean'),
    total_estimated_revenue=('estimated_annual_revenue', 'sum'),
    avg_estimated_revenue=('estimated_annual_revenue', 'mean')
).round(2)
revenue_by_area
```

```python
"""
CUSTOM FUNCTIONS — .apply()
- .apply(function) runs a function on every value in a column (or every row).
- Here we create a custom function 'price_tier' that classifies each price into a category.
- This is like investment grade ratings (AAA, AA, A, etc.) but for pricing.
- .apply() is flexible but slower than vectorized operations for large datasets.
- Alternative for simple conditions: np.where() or pd.cut() (both are faster).
"""
def price_tier(price):
    if price < 50:
        return 'Budget'
    elif price < 100:
        return 'Economy'
    elif price < 200:
        return 'Mid-Range'
    elif price < 500:
        return 'Premium'
    else:
        return 'Luxury'

data['price_tier'] = data['price'].apply(price_tier)
print("=== Listing Distribution by Price Tier ===")
print(data['price_tier'].value_counts())
print("\n=== Average Revenue by Price Tier ===")
data.groupby('price_tier')['estimated_annual_revenue'].mean().round(2)
```

```python
"""
RANKING DATA — .rank()
- .rank() assigns a rank (position) to each value in a column.
- ascending=False: highest value gets rank 1 (e.g., most expensive = rank 1).
- method='dense': no gaps in ranking (1, 2, 3 not 1, 2, 4 if there's a tie).
- Other methods: 'average' (default), 'min', 'max', 'first' (by row order).
- Financial analysts use ranking for: leaderboards, competitive positioning, and percentile analysis.
"""
data['price_rank'] = data['price'].rank(ascending=False, method='dense').astype(int)
data['revenue_rank'] = data['estimated_annual_revenue'].rank(ascending=False, method='dense').astype(int)
print("=== Top 10 by Revenue Rank ===")
data.nsmallest(10, 'revenue_rank')[['name', 'price', 'revenue_rank', 'estimated_annual_revenue']]
```

```python
"""
BINNING DATA — pd.cut() and pd.qcut()
- pd.cut(column, bins, labels): creates bins with EQUAL WIDTH (you define the boundaries).
  Example: bins=[0,50,100,200] creates ranges 0-50, 50-100, 100-200.
- pd.qcut(column, q, labels): creates bins with EQUAL FREQUENCY (same number of items per bin).
  Example: q=4 creates 4 bins, each with ~25% of the data (quartiles).
- Use cut() when you have meaningful boundaries (e.g., price ranges).
- Use qcut() when you want equal-sized groups (e.g., quartile analysis).
- Labels make the output human-readable instead of showing interval notation.
"""
data['price_bin'] = pd.cut(data['price'], bins=[0, 50, 100, 200, 500, 10000],
                           labels=['$0-50', '$50-100', '$100-200', '$200-500', '$500+'])
data['price_quartile'] = pd.qcut(data['price'], q=4, labels=['Q1-Cheapest', 'Q2', 'Q3', 'Q4-Priciest'])

print("=== Price Bin Distribution ===")
print(data['price_bin'].value_counts().sort_index())
print("\n=== Price Quartile Distribution ===")
print(data['price_quartile'].value_counts().sort_index())
```

```python
"""
STRING OPERATIONS — .str accessor
- .str gives access to string methods on text columns.
- .str.contains('pattern') returns True/False — great for searching text.
  - case=False: case-insensitive search
  - na=False: treat missing values as False (not a match)
  - Use | for OR: 'luxury|modern|beautiful'
- Other useful .str methods:
  - .str.upper() / .str.lower(): change case
  - .str.len(): length of each string
  - .str.split(): split text into parts
  - .str.replace(): find and replace text
  - .str.startswith() / .str.endswith(): check beginning/end of text
"""
luxury = data[data['name'].str.contains('luxury|modern|beautiful', case=False, na=False)]
print(f"Listings with 'luxury/modern/beautiful' in name: {len(luxury)}")
print(f"Average price of these listings: ${luxury['price'].mean():.2f}")
print(f"Average price of ALL listings:   ${data['price'].mean():.2f}")

data['name_length'] = data['name'].str.len()
print(f"\nAverage listing name length: {data['name_length'].mean():.0f} characters")
```

## 7. Data Visualization — Charts That Tell the Story
Pandas has built-in plotting powered by matplotlib. A financial analyst presents findings visually.

```python
"""
IMPORT MATPLOTLIB — Visualization library
- matplotlib.pyplot is Python's most popular plotting library.
- We import it as 'plt' for short.
- plt.style.use() sets a visual theme for all charts.
- plt.rcParams sets default figure size (width, height) in inches.
"""
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
"""
HISTOGRAM — .plot.hist()
- A histogram shows the frequency distribution of a numeric column.
- bins=50 splits the data into 50 equal-width bars.
- alpha=0.7 makes bars slightly transparent.
- plt.axvline() draws a vertical reference line (here: the median price).
- Useful for: understanding price distribution, spotting skewness, identifying outliers.
"""
data[data['price'] < 500]['price'].plot.hist(bins=50, edgecolor='black', alpha=0.7)
plt.title('Price Distribution of Seattle Listings (under $500)')
plt.xlabel('Price ($)')
plt.ylabel('Number of Listings')
plt.axvline(data['price'].median(), color='red', linestyle='--', label=f'Median: ${data["price"].median():.0f}')
plt.legend()
plt.show()
```

```python
"""
BAR CHART — .plot.bar()
- Bar charts compare values across categories (e.g., average price per room type).
- .sort_values() orders bars from smallest to largest for easy comparison.
- plt.xticks(rotation=45) rotates x-axis labels so they don't overlap.
- plt.tight_layout() prevents labels from being cut off.
"""
data.groupby('room_type')['price'].mean().sort_values().plot.bar(
    color=['#2ecc71', '#3498db', '#e74c3c', '#f39c12'], edgecolor='black')
plt.title('Average Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
"""
HORIZONTAL BAR CHART — .plot.barh()
- Same as bar chart but horizontal — better when category names are long.
- .value_counts().head(10) gets the top 10 most common neighbourhoods.
- Useful for: ranking categories, showing top/bottom performers.
"""
data['neighbourhood'].value_counts().head(10).plot.barh(color='#3498db', edgecolor='black')
plt.title('Top 10 Neighbourhoods by Number of Listings')
plt.xlabel('Number of Listings')
plt.ylabel('Neighbourhood')
plt.tight_layout()
plt.show()
```

```python
"""
PIE CHART — .plot.pie()
- Pie charts show proportions / market share of categories.
- autopct='%1.1f%%' displays percentage labels on each slice.
- startangle=90 rotates the chart so the first slice starts at 12 o'clock.
- Best for: showing how a whole is divided into parts (e.g., room type market share).
- Avoid for >6 categories — bar charts are easier to read.
"""
data['room_type'].value_counts().plot.pie(
    autopct='%1.1f%%', startangle=90,
    colors=['#2ecc71', '#3498db', '#e74c3c', '#f39c12'])
plt.title('Market Share by Room Type')
plt.ylabel('')
plt.show()
```

```python
"""
BOX PLOT — .boxplot()
- Box plots show the distribution of data with 5 key values:
  - Box: 25th to 75th percentile (middle 50% of data)
  - Line inside box: median (50th percentile)
  - Whiskers: extend to 1.5x the box height
  - Dots beyond whiskers: outliers (unusually high/low values)
- Great for comparing distributions across categories and spotting outliers.
"""
data[data['price'] < 500].boxplot(column='price', by='room_type', figsize=(10, 6))
plt.title('Price Distribution by Room Type')
plt.suptitle('')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.show()
```

```python
"""
SCATTER PLOT — .plot.scatter()
- Scatter plots show the relationship between two numeric variables.
- Each dot = one listing, positioned by its price (x) and reviews (y).
- alpha=0.3 makes dots semi-transparent so overlapping points are visible.
- Look for patterns: clusters, trends (upward/downward), or no relationship.
- In finance, scatter plots help visualize correlations between metrics.
"""
data[data['price'] < 500].plot.scatter(x='price', y='number_of_reviews', alpha=0.3, color='#e74c3c')
plt.title('Price vs Number of Reviews')
plt.xlabel('Price ($)')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
"""
STACKED BAR CHART — .plot.bar(stacked=True)
- Stacked bars show how sub-categories contribute to each group's total.
- Each color = a room type; total bar height = total revenue for that area.
- .unstack() pivots room_type from rows into columns (needed for stacking).
- bbox_to_anchor=(1.05, 1) places the legend outside the chart.
- Great for: revenue breakdown, market composition, segment contribution analysis.
"""
revenue_pivot = data.groupby(['neighbourhood_group', 'room_type'])['estimated_annual_revenue'].sum().unstack()
revenue_pivot.plot.bar(stacked=True, figsize=(10, 6), colormap='Set2', edgecolor='black')
plt.title('Total Estimated Revenue by Area & Room Type')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Total Estimated Revenue ($)')
plt.legend(title='Room Type', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```
